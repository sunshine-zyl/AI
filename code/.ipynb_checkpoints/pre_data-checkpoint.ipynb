```python
# import packages
import numpy as np
import pandas as pd
```

```python
# data path
src_path = "~/MyProject/data/"
file_name = "cpu_8.xls"
file_path = src_path + file_name
```

```python
# read data
df = pd.read_excel(file_path)
df.head()
```

```python
df.shape
```

```python
# messages
df.describe()
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
dv = df["cpu"][0:672]
dv
```

```python
res = sm.tsa.seasonal_decompose(dv, period=96)

res.plot()
plt.show()
```

```python
def my_stl(file_path, freq=96):
    df = pd.read_excel(file_path)
    dv = df["cpu"][0:672]
    res = sm.tsa.seasonal_decompose(dv, period=freq)
    res.plot()
    plt.show()
```

```python
# cpu_18.xls
my_stl(src_path + "cpu_18.xls")
```

```python
# cpu_9.xls
my_stl(src_path + "cpu_9.xls")
```
